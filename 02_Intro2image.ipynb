```python
# Load the image
data = np.array(plt.imread('./images/sign.jpg'))


# Set the red channel in this part of the image to 0
#yaeni az satre 0 ta satre 1000, sotune 0 ta sotune 1000 boro redesho bokon 0 yaeni red ndashte bashe
data[0:1000, 0:1000, 0] = 0

# Set the green channel in this part of the image to 255
#hala ke redesho 0 kardikm sabzesho ta jayi ke hast ziad kon
data[0:1000, 0:1000, 1] = 255

# Set the blue channel in this part of the image to 0
#hala ke redesho gerefti sabzesho ziad kardi abisham 0 kon
data[0:1000, 0:1000, 2] = 0

# Visualize the result
plt.imshow(data)
plt.show()
```

## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma"> تصاویر سطح خاکستری  </div>


```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114]).astype(int)
```

<div style="direction:rtl;text-align:right;font-family:Tahoma">
خواندن یک تصویر به صورت سطح خاکستری</div>

```python
img = plt.imread('./images/t-shirt.jpg')
tshirt = rgb2gray(img)
```

<div style="direction:rtl;text-align:right;font-family:Tahoma">
اندازه ماتریس خوانده شده</div>

```python
tshirt.shape
```

<div style="direction:rtl;text-align:right;font-family:Tahoma">
نمایش تصویر</div>

```python
plt.imshow(tshirt, cmap = plt.get_cmap('gray'))
plt.show()
```
